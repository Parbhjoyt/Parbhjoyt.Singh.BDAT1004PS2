5.Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.

```python
#  Code 5

def binary(n):
    if(n==1):
        print(1,end='')
    elif(n>1):
        binary(n//2)
        print(n%2,end='') 
    elif(n==0):
        print(0)
    
binary(9)
```

7.Implement recursive function webdir() that takes as input: a URL (as a string) and

each web page's URL. As shown below, indentation, specified by indent, should
be used to indicate the depth of a URL

```python
# Code 7
from urllib.request import urlopen
from urllib.parse import urljoin
from html.parser import HTMLParser

def webdir(url,depth=0,indent=0):
    inn = indent
    dep = depth
    global visited_url    
    visited_url.add(url)
    if url in visited_url:
        if indent==0:
            print(url)
        elif indent==1:
            print(" ",url) 
        elif indent==2:
            print("  ",url)

    web_c = urlopen(url).read().decode()
    c = urlParser(url)
    c.feed(web_c)
    url_links = c.geturl() 

    for l in url_links:

        if l in visited_url:
            if inn==0:
                print(" ",l)
            elif inn==1:
                print("  ",l) 
            elif inn==2:
                print("   ",l)
        else:
            if(dep>0):  
                x=inn+1
                y=dep-1
                webdir(l,y,x)
                
class urlParser(HTMLParser):
    def __init__(self, url):
        HTMLParser.__init__(self)
        self.url = url
        self.url_links = []
    
    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for a in attrs:
                if a[0] == 'href':
                    u_join = urljoin(self.url, a[1])
                    if u_join[0:4] == 'http':
                        self.url_links.append(u_join)
            
    def geturl(self):
        return self.url_links

visited_url = set()

webdir('http://reed.cs.depaul.edu/lperkovic/test1.html',2,0)
```

and 400 millimeters.
f) The city and country for which the average Fall temperature is above 20
degrees, in increasing temperature order.
g) The total annual rainfall for Cairo.
h) The total rainfall for each season.

```python
# Code 8

import sqlite3
con = sqlite3.connect('SQLDB.db')
cur = con.cursor()


# a) All the temperature data
cur.execute('SELECT Temperature FROM SQLDBTABLE')
cur.fetchall()
```

```python
#b) All the cities, but without repetition.
cur.execute('SELECT DISTINCT(City) FROM SQLDBTABLE')
cur.fetchall()
```

```python
#c) All the records for India.
cur.execute('SELECT * FROM SQLDBTABLE WHERE Country = "India" ')
cur.fetchall()
```

```python
#d) All the Fall records.
cur.execute('SELECT * FROM SQLDBTABLE WHERE Season = "Fall" ')
cur.fetchall()
```

```python
#e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters
cur.execute('SELECT City,Country,Season FROM SQLDBTABLE WHERE Rainfall>200 AND Rainfall<400 ')
cur.fetchall()
```

```python
#f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.
cur.execute('SELECT City,Country FROM SQLDBTABLE WHERE Season="Fall" AND Temperature>20 ORDER BY Temperature ASC  ')
cur.fetchall()
```

```python
#g) The total annual rainfall for Cairo.
cur.execute('SELECT SUM(Rainfall) FROM SQLDBTABLE WHERE City="Cairo"')
cur.fetchall()
```

```python
#h) The total rainfall for each season.
cur.execute('SELECT Season,SUM(Rainfall) FROM SQLDBTABLE GROUP BY Season')
cur.fetchall()
```
